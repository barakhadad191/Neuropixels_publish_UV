Load the files dicts

```python
def load_file_dict(filename):
    # Dictionary to store the results
    svm_results = {}
    
    with open(filename, "r") as file:
        for line in file:
            # Split the line to extract layer and accuracy
            parts = line.strip().split(", ")
            layer_name = parts[0].split(": ")[1]
            accuracy = float(parts[1].split(": ")[1].replace("%", ""))
            
            # Store in dictionary
            svm_results[layer_name] = accuracy
    return svm_results
```

```python
main_folder = r'\\132.66.34.210\Pixel1\1_auditory_neuropixels_BarakH\Analyzed_C9'
low_path = main_folder + r"\APmodel_SimpleCNN_100_199_acc89.60_svm_accuracies.txt"
high_path = main_folder + r"\APmodel_SimpleCNN_0_99_acc65.70_svm_accuracies.txt"

results_low = load_file_dict(low_path)
results_high = load_file_dict(high_path)
```

```python
import matplotlib.pyplot as plt

custom_labels = [f"conv{i+1}" if "conv" in layer else f"fc{i-4}" for i, layer in enumerate(results_low.keys())]

plt.figure(figsize=(10, 6))
plt.plot(custom_labels, results_low.values(), color='skyblue', label='Low')
plt.plot(custom_labels, results_high.values(), color='orange', label='High')
plt.xlabel('Accuracy')
plt.title('Layer-wise SVM Accuracies')
plt.grid(axis='x', linestyle='--')

# Show the plot
plt.tight_layout()
plt.show()
```
